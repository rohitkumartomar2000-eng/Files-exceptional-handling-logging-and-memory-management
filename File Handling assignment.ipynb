```python
# Q15 How does Python's garbage collection system work ?

#Answer:-
# Python’s garbage collection system is responsible for automatically removing unused objects from memory so that memory can be reused efficiently. 
# The programmer does not need to manually delete memory.

# 1. Reference Counting

# Every object in Python keeps a count of how many references point to it.
# When a variable is assigned to an object, the reference count increases.
# When a variable is deleted or reassigned, the reference count decreases.
# When the reference count becomes zero, Python immediately frees that object from memory.

# 2. Problem of Circular References

# Sometimes objects reference each other (for example, object A refers to B and B refers to A).
# In such cases, reference counts never become zero, even though the objects are useless.
# This is where simple reference counting fails.

# 3. Cyclic Garbage Collector

# To handle circular references, Python uses a cyclic garbage collector.
# It periodically scans objects in memory and identifies groups of objects that reference each other but are no longer reachable by the program.
# These unused objects are then safely removed.

# 4. Automatic and Background Process

# Garbage collection runs automatically in the background.
# The programmer does not need to call it manually in normal situations.
# This keeps memory usage under control and improves program stability.

# Why this system matters

# Prevents memory wastage
# Avoids memory leaks
# Improves performance of long-running programs
    
# conclusion

# Python’s garbage collection system works by using reference counting and a cyclic garbage collector to automatically detect 
# and remove unused objects from memory, ensuring efficient memory management without manual intervention
```

```python
# Q18 . What is the difference between os.fork() and multiprocessing in Python ?

#Answer:-
# In Python, both os.fork() and the multiprocessing module are used to run tasks in separate processes,
# but they are not designed for the same purpose and should not be treated as equals

# The os.fork() method works at the operating system level. It creates a new process by copying the currently running process.
# After forking, both the parent and child processes continue running independently. 
# However, this method is restricted to Unix-based systems like Linux and macOS and is not available on Windows.
# Managing processes using os.fork() requires more care and deeper system-level understanding.

# On the other hand, the multiprocessing module is a Python-provided solution that simplifies process creation.
# It allows a program to run multiple processes easily while keeping them isolated from each other. 
# Unlike os.fork(), it works on all major operating systems, including Windows. 
# It also provides built-in tools to manage processes, making it safer and more suitable for real-world Python applications.

# Main Differences at a Glance

# os.fork() is low-level and system-dependent
# multiprocessing is high-level and cross-platform
# os.fork() requires manual process handling
# multiprocessing handles most complexity automatically
# multiprocessing is preferred for modern Python programs

#  conclusion

# os.fork() directly creates a child process by duplicating the parent process and is limited to Unix systems, 
# while the multiprocessing module offers a higher-level, portable, and safer way to create and manage multiple processes in Python.
```

```python
# #Q23 What are the challenges associated with memory management in Python ?

# #Answer:- 
# Although Python handles memory automatically, memory management still comes with several challenges that programmers must be aware of.

# 1. Memory leaks

# Even with automatic memory management, memory leaks can occur if objects are still referenced unnecessarily.
# When references are not removed properly, Python cannot free the memory, which leads to increased memory usage.

# 2. Circular references

# Sometimes objects reference each other in a loop. 
# These circular references cannot be cleaned up easily using reference counting alone and depend on garbage collection, which may not run immediately.

# 3. High memory consumption

# Python objects consume more memory compared to lower-level languages. 
# Large data structures, unused variables, or inefficient coding can cause programs to use excessive memory.

# 4. Delayed garbage collection

# Garbage collection does not always run instantly. This can cause unused memory to stay allocated longer than expected, 
# especially in long-running applications.

# 5. Difficulty in controlling memory manually

# Python does not give direct control over memory allocation and deallocation. While this simplifies coding, 
# it makes memory optimization harder in performance-critical applications.

#  conclusion

# The main challenges of memory management in Python include memory leaks, circular references, high memory usage, delayed garbage collection, 
# and limited manual control over memory, which can affect performance if not handled carefully.
```

```python
#Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB ?

import logging
from logging.handlers import RotatingFileHandler

logger = logging.getLogger()
logger.setLevel(logging.INFO)

handler = RotatingFileHandler(
    "app.log",
    maxBytes=1_000_000,
    backupCount=3
)

formatter = logging.Formatter(
    "%(asctime)s - %(levelname)s - %(message)s"
)
handler.setFormatter(formatter)

logger.addHandler(handler)

logger.info("Application started")
logger.warning("This is a warning message")
logger.error("This is an error message")
```

```python
#Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word ?

word = "python"
count = 0

with open("data.txt", "r") as file:
    for line in file:
        count += line.lower().split().count(word.lower())

print("Number of occurrences:", count)
```

```python
#Q22. How can you check if a file is empty before attempting to read its contents ?

import os

file_name = "data.txt"

if os.path.exists(file_name) and os.path.getsize(file_name) == 0:
    print("The file is empty.")
else:
    with open(file_name, "r") as file:
        print(file.read())
```

```python
#Q23. Write a Python program that writes to a log file when an error occurs during file handling.

import logging

logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    with open("data.txt", "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    logging.error("File not found while attempting to read data.txt")
    print("Error: File not found.")

except Exception as e:
    logging.error(f"Unexpected error occurred: {e}")
    print("Error: Something went wrong.")
```
